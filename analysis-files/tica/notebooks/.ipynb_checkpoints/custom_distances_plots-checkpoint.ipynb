```python
import pyemma as py
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt
import itertools
import seaborn as sns
```

```python
# import distances:
load_path = '../'
# WT:
which_system = '1m47_c125s'
distances_data_wt = np.load(f'{load_path}data/system_{which_system}_distances_custom_v1.npy', allow_pickle=True)
WT_resis     = np.load(f'{load_path}data/system_{which_system}_distances_custom_v1_resis.npy', allow_pickle=True)

pdb_path = f'/Users/ssolieva/Desktop/bowman_lab/projects/IL2/simulations/IL2_simulations_{which_system}/{which_system}-prot-masses.pdb'
top_file = md.load(pdb_path) # load in the pdb file

# S15
which_system = 'seq15_c125s'
distances_data_s15 = np.load(f'{load_path}data/system_{which_system}_distances_custom_v1.npy', allow_pickle=True)
S15_resis     = np.load(f'{load_path}data/system_{which_system}_distances_custom_v1_resis.npy', allow_pickle=True)

# S1
which_system = 'seq1_c123s'
distances_data_s1 = np.load(f'{load_path}data/system_{which_system}_distances_custom_v1.npy', allow_pickle=True)
S1_resis     = np.load(f'{load_path}data/system_{which_system}_distances_custom_v1_resis.npy', allow_pickle=True)

# L56A
which_system = 'seq1_c123s_L56A'
distances_data_L56A = np.load(f'{load_path}data/system_{which_system}_distances_custom_v1.npy', allow_pickle=True)
L56A_resis     = np.load(f'{load_path}data/system_{which_system}_distances_custom_v1_resis.npy', allow_pickle=True)

# V84A
which_system = 'seq1_c123s_V84A'
distances_data_V84A = np.load(f'{load_path}data/system_{which_system}_distances_custom_v1.npy', allow_pickle=True)
V84A_resis     = np.load(f'{load_path}data/system_{which_system}_distances_custom_v1_resis.npy', allow_pickle=True)

print(len(distances_data_s1), len(distances_data_s15))
```

```python
# format data for analysis:
r2_wt   = np.transpose(np.concatenate(distances_data_wt, axis=0))
r2_s15  = np.transpose(np.concatenate(distances_data_s15, axis=0))
r2_s1   = np.transpose(np.concatenate(distances_data_s1,  axis=0))
r2_L56A = np.transpose(np.concatenate(distances_data_L56A,  axis=0))
r2_V84A = np.transpose(np.concatenate(distances_data_V84A,  axis=0))
```

```python
# resi pairs:
IL_resin =  '17 23 26 82 85 89 122'
atom_indices_IL = top_file.topology.select(f'residue {IL_resin} and name CA')
atom_indices1 = np.unique(np.concatenate([atom_indices_IL])) # put in order
atom_indices = np.unique(atom_indices1)
atom_pairs = list(itertools.combinations(atom_indices, 2))
print(len(atom_pairs))
resi_n_names = []
for i in range(len(atom_indices1)):
    resi_n_names.append(top_file.topology.atom(atom_indices[i]).residue)
print(resi_n_names)
distances_feat = py.coordinates.featurizer(top_file)
distances_feat.add_distances(atom_pairs)
```

```python
def plot_dist(n):
    WT_n   =  r2_wt[n]
    S15_n  = r2_s15[n] 
    S1_n   =  r2_s1[n]
    L56A_n =  r2_L56A[n]
    V84A_n =  r2_V84A[n]
    
    plt.figure(figsize=[6,3])
    angstrom = r'$\bf{\AA}$'
    plt.xlabel(f"{list(distances_feat.describe())[n]}", fontsize=14, weight='bold')
    plt.ylabel('Density',fontsize=14, weight='bold')
    sns.kdeplot(WT_n*10,    c='#888985', label='WT',   linewidth=5)
    sns.kdeplot(S15_n*10,   c='#5AC8E0', label='S15',  linewidth=5)
    sns.kdeplot(S1_n*10,    c='#BC60A5', label='S1',   linewidth=5)
    sns.kdeplot(L56A_n*10,  c='lightgray', label='L56A',   linewidth=5)
    sns.kdeplot(V84A_n*10,  c='lightpink', label='V84A',   linewidth=5)
    plt.xticks(fontsize=12, weight='bold')
    plt.yticks(fontsize=12, weight='bold')
    plt.legend(fontsize=8)
    plt.tight_layout()
    #plt.savefig(f'../figures/distance_distributions/c125s_pair_{n}.pdf', transparent=True, dpi=300)
    plt.show()
```

```python
plot_dist(2)
plot_dist(3)
plot_dist(4)
plot_dist(5)
plot_dist(11)
```
