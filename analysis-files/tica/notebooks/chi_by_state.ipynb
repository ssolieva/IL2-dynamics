```python
def plot_nmr_vs_md(mdsimRes, nmrRes, MD_data, NMR_data, NMR_color, title):
    plt.figure(figsize=[6,4])
    plt.plot(nmrRes, NMR_data, marker='o', markersize=2, linewidth=1, c=NMR_color, label='NMR')
    plt.plot(mdsimRes, MD_data,  marker='o', markersize=2, linewidth=1, c='k', label='MD')
    plt.ylabel("Amide Order Parameter", fontsize=16, weight='bold')
    plt.xlabel("Residue", fontsize=16, weight='bold')
    plt.legend(fontsize=12)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.ylim(0,1)
    plt.title(f'{title}', fontsize=16, weight='bold')
    plt.tight_layout()
    plt.show()
```
