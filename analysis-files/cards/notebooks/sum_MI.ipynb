```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_sum_MI_plot(rr_mat, color, system, short):
    norm_sum_MI = []
    for i in range(len(rr_mat)):
        norm_sum_MI.append(sum(rr_mat[i]))
    
    resi_list = []
    if short==True:
        for i in range(len(np.arange(len(rr_mat)))):
            if i <= 80-1:
                resi_list.append(np.arange(1,len(rr_mat)+1)[i])
            if i > 80-1:
                resi_list.append(np.arange(1,len(rr_mat)+1)[i]+2)
    if short==False:
        resi_list = np.arange(1,len(rr_mat)+1)
    #print(resi_list)
    #print(len(resi_list))

    plt.figure(figsize=[12,2])
    plt.bar(resi_list, norm_sum_MI, width=0.7, color=color, edgecolor='k')
    plt.xticks(fontsize=9, weight='bold')
    plt.ylim(0,30)
    plt.ylabel('normalized\nsum(MI)', fontsize=14, weight='bold')
    plt.yticks(fontsize=10, weight='bold')
    plt.xlabel('residue', fontsize=14, weight='bold')
    plt.title(f"{system}", fontsize=14, weight='bold')
    #plt.xlim(20,70)
    plt.tight_layout()
    
    plt.savefig(f'../figures/full_sum_MI_{system}.pdf', dpi=300, transparent=True)
    return resi_list, norm_sum_MI
```

```python
# WT: 
rr_mat = np.genfromtxt('../data/1m47_rr-mi.dat',skip_header=0,skip_footer=0,dtype=None,delimiter=' ')
resi_list, norm_sum_MI = generate_sum_MI_plot(rr_mat, '#888985', 'WT', False)
np.savetxt('WT_resi_list.txt', resi_list, delimiter=',')
np.savetxt('WT_sum_MI.txt', norm_sum_MI, delimiter=',')

rr_mat = np.genfromtxt('../data/seq1_c123s_rr-mi.dat',skip_header=0,skip_footer=0,dtype=None,delimiter=' ')
resi_list, norm_sum_MI = generate_sum_MI_plot(rr_mat, '#BC60A5', 'S1', True)
np.savetxt('S1_resi_list.txt', resi_list, delimiter=',')
np.savetxt('S1_sum_MI.txt', norm_sum_MI, delimiter=',')

rr_mat = np.genfromtxt('../data/seq15_rr-mi.dat',skip_header=0,skip_footer=0,dtype=None,delimiter=' ')
resi_list, norm_sum_MI = generate_sum_MI_plot(rr_mat, '#5AC8E0', 'S15', False)
np.savetxt('S15_resi_list.txt', resi_list, delimiter=',')
np.savetxt('S15_sum_MI.txt', norm_sum_MI, delimiter=',')

rr_mat = np.genfromtxt('../data/seq1_c123s_L56A_rr-mi.dat',skip_header=0,skip_footer=0,dtype=None,delimiter=' ')
resi_list, norm_sum_MI = generate_sum_MI_plot(rr_mat, 'goldenrod', 'S1_L56A', True)

#rr_mat = np.genfromtxt('../data/seq1_c123s_V84A_rr-mi.dat',skip_header=0,skip_footer=0,dtype=None,delimiter=' ')
#resi_list, norm_sum_MI = generate_sum_MI_plot(rr_mat, 'lightgray', 'S1_V84A', True)
```
